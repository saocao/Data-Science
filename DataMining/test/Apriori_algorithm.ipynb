```python
import numpy as np
f = open('categories.txt','r')
```

```python
f1 = f.readlines()
li_data = [] # contains data
for x in f1:
    li_data.append(sorted(x[:-1].split(';')))
```

```python
flat_data = sorted(set([val for sublist in li_data for val in sublist]))
flat_data
```

```python
flat_data = [1, 2,3,4,5]
li_data = [[1,2],[2,3],[3,5]]
onehot_data = [[0]*len(flat_data)]*len(li_data)
idx_item = 0
for i in range(len(li_data)):
    for item in li_data[i]:
        idx_item = flat_data.index(item)
        onehot_data[i][idx_item] = 1
```
